# Importing Packages

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

# Reading and Normalizing Data

```python
data = np.loadtxt('logreg_data_binary.txt')
n_obj = data.shape[0]
n_col = data.shape[1]
labels = data[:, 0]
features = data[:, np.arange(1, data.shape[1])]
raw_features = features
```

```python
mean_features = np.mean(features, axis=0)
rev_std_features = [1.0/x for x in np.std(features, axis=0)]
features = [np.multiply(x - mean_features, rev_std_features) for x in features]
features = np.hstack((np.ones(n_obj, ).reshape(n_obj, 1), features))
n_features = features.shape[1]
```

# NLL and its gradient

```python
def LogReg_prob(w, x):
    return 1/(1 + np.exp(-np.dot(x, w)))
```

```python
def NLL(features, labels, w):
    res = 0
    n_obj = features.shape[0]
    for i in range(n_obj):
        res -= labels[i] * np.log(LogReg_prob(w, features[i])) + (1 - labels[i]) * np.log(1 - LogReg_prob(w, features[i]))
    return res
```

```python
def NLL_grad(features, labels, w):
    grad = np.zeros_like(features[0])
    n_obj = features.shape[0]
    for i in range(n_obj):
        grad += (LogReg_prob(w, features[i]) - labels[i])*features[i]
    return grad
```

# Task 1

Let us draw the graph of NLL decreasing. Stepsize of the GD is $\frac{1}{\text{number  of objects}}$.

```python
n_iter = 50
f_values = np.zeros((n_iter, ))
step_size = 1/n_obj
x_k = np.zeros(n_features)
```

```python
for i in range(n_iter):
    f_values[i] = NLL(features, labels, x_k)
    x_k -= step_size * NLL_grad(features, labels, x_k)
```

```python
plt.figure(figsize = (10, 10))
plt.plot(f_values, label="NLL")
plt.xlim([0, n_iter])
plt.xlabel("iteration number")
plt.grid()
plt.ylabel("NLL")
plt.savefig("NLL_decreasing.eps")
```
