# Importing libraries

```python
import sklearn
from matplotlib import pyplot as plt
import numpy as np
import scipy
import pandas as pd
from sklearn.svm import SVC
from scipy import linalg
from sklearn import preprocessing
```

# Loading Data, Basic Preprocessing

```python
data = pd.read_csv("../HW_7_SVM/winequality-red.csv", sep=';')
data['class'] = -1
data.loc[data.quality >= 6, 'class'] = 1
features, labels = data.drop(['class', 'quality'], axis = 1), data['class']
n_obj = features.shape[0]
```

```python
t_ind = int(n_obj * 0.6)
x_train = features[:t_ind]
y_train = labels[:t_ind]
x_test = features[t_ind:]
y_test = labels[t_ind:]
```

# PCA transformation

```python
def pca(x, m):
    n = x.shape[0]
    mean = np.mean(x, axis = 0)
    S = (x - mean).T @ (x - mean) / n
    eig = linalg.eigh(S)
    eigv = eig[1][-m:].T
    return eigv
```

# SVM

```python
def scores_svm(train_set, test_set):
    model = SVC(kernel = 'linear').fit(train_set, y_train)
    scores = model.decision_function(test_set)
    return scores
```

# SVM with PCA

```python
n = 0
# ks = np.arange(1, 12)
ks = [1, 4, 8, 11]
fig, ax = plt.subplots(1, len(ks), sharey=True, figsize = (30, 10))
auc_roc = dict()
for k in ks:
    eigv = pca(x_train, k)
    scores = scores_svm(x_train @ eigv, x_test @ eigv)
    fpr, tpr, _ = sklearn.metrics.roc_curve(y_test, scores)
    auc_roc[k] = sklearn.metrics.roc_auc_score(y_test, scores)
    ax[n].set_title("ROC : {}".format(k))
    ax[n].plot(fpr, tpr)
    n += 1
# plt.savefig("ROC_pca.png")
```

# SVM with SVD

```python
n = 0
# ks = np.arange(1, 12)
ks = [1, 4, 8, 11]
fig, ax = plt.subplots(1, len(ks), sharey=True, figsize = (30, 10))
auc_roc = dict()
for k in ks:
    u, s, vh = np.linalg.svd(features, full_matrices=True)
    new_features = u[:, :k] @ np.diag(s[:k]) @ vh[:k, :]
    scores = scores_svm(new_features[:t_ind], new_features[t_ind:])
    fpr, tpr, _ = sklearn.metrics.roc_curve(y_test, scores)
    auc_roc[k] = sklearn.metrics.roc_auc_score(y_test, scores)
    ax[n].set_title("ROC : {}".format(k))
    ax[n].plot(fpr, tpr)
    n += 1
# plt.savefig("ROC_svd.png")
```
