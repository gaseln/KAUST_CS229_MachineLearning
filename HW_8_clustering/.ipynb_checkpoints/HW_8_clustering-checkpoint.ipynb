# Importing libraries

```python
import numpy as np
import scipy
from matplotlib import pyplot as plt
from numpy import linalg as LA
import seaborn as sns
import pandas as pd
```

# Loading data, initial plot

```python
data = np.loadtxt("clu_data.txt")
```

```python
plt.scatter(data[:, 0], data[:, 1], s=2)
```

# K-means

```python
def kmeans(data:np.ndarray, k, max_iter=1000, tolerance=1e-3):
    x = data[:, 0]
    y = data[:, 1]
    n_points = len(x)
    max_x = np.max(x)
    min_x = np.min(x)
    max_y = np.max(y)
    min_y = np.min(y)
    centers = np.stack([min_x + (max_x - min_x) * np.random.random_sample(k), 
                        min_y + (max_y - min_y) * np.random.random_sample(k)], axis=1)
    distances = np.zeros((n_points, k))
    classes = np.zeros(n_points)
    indicator = np.zeros(n_points)
    halt_flag = True
    for i in range(max_iter):
        halt_flag = True
        for j in range(k):
            distances[:, j] = ((data - centers[j])**2).sum(axis=1)
        classes = np.argmin(distances, axis=1)
        for j in range(k):
            indicator = (classes == j).astype(int)[:, None]
            assert indicator.sum() > 0
            center = (indicator * data).sum(axis=0) / indicator.sum()
            if LA.norm(center - centers[j]) > tolerance:
                halt_flag = False
            centers[j] = center
        if halt_flag:
            break
    SSE = np.min(distances, axis=1).sum() / n_points
    return centers, SSE, classes
```

# SSE plot

```python
max_k = 15
SSE = np.zeros(max_k, tolerance=1e-2)
for k in range(max_k):
    _, SSE[k], _ = kmeans(data, k + 1) # k starts from zero
```

```python
plt.figure(figsize=(10, 10))
plt.title("SSE")
plt.xlabel("number of clusters")
plt.ylabel("SSE")
plt.plot(range(1, max_k + 1), SSE)
plt.savefig("SSE.png")
```

# Manually choosing best k

```python
best_k = 4
_, _, classes = kmeans(data, best_k, max_iter=10000, tolerance=1e-20)
classes = classes[:, None]
```

```python
df = pd.DataFrame(data, columns={'x', 'y'})
df['class'] = classes
```

```python
facet = sns.lmplot(data=df, x='x', y='y', hue='class', 
                   fit_reg=False, legend=True, legend_out=True)
facet.savefig("k-means.png")
```
