# Sequential Bayesian Learning

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import multivariate_normal
%matplotlib inline
```

```python
# covariance matrix of posterior Gaussian distribution
def gauss_cov(β, Φ, S_0):
    return np.inv(np.inv(S_0) + β * np.dot(np.transpose(Φ), Φ))
```

```python
# mean of posterior Gaussian distribution
def gauss_mean(β, Φ, S_0, t, m_0):
    return np.dot(gauss_cov(β, Φ, S_0), β * np.dot(np.transpose(Φ), t) + np.dot(np.inv(S_0), m_0))
```

```python
def draw_2D_gaussian(ax, mean, cov):
    [x, y] = np.mgrid[-1:1:.01, -1:1:.01]
    pos = np.dstack((x, y))
    gauss = multivariate_normal(mean, cov)
    ax.contourf(x, y, gauss.pdf(pos))
```

```python
draw_2D_gaussian([0.3, 0.2], [[0.01, 0], [0, 0.01]])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (30, 10))
draw_2D_gaussian(ax1, [0.3, 0.2], [[0.01, 0], [0, 0.01]])
draw_2D_gaussian(ax2, [0.3, 0.3], [[0.01, 0], [0, 0.01]])
plt.draw()
```
